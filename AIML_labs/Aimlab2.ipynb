# Problem Statement:
Develop a linear regression model to predict sales revenue based on the advertising budget. The model should help businesses identify how much investment in advertising is generally required to achieve a certain level of sales, thereby aiding in the development of effective marketing strategies and budget allocations.


```python
# Importing libraries required for plotting and other necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# Reading the csv file from Desktop
df = pd.read_csv(r"C:\Users\sayed\Desktop\Practice_Dataset\Lab-1\business_sales_data.csv")
print(df.head())
```

```python
# Scatter plot of Advertising Budget and Sales
plt.scatter(df['Advertising_Budget'], df['Sales'])
plt.title('Advertising Budget vs Sales')
plt.xlabel('Advertising Budget')
plt.ylabel('Sales')
plt.show()
```

```python
# Correlation check
correlation = df.corr()
correlation
```

```python
df.head()
```

```python
# Taking independent and dependent variables into consideration 
X = df[['Advertising_Budget']]   
Y =df['Sales'] 
```

```python
# Training model 80 percent train and 20 percent for testing 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=0)
```

```python
# Printing the result from the model
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
X_train.head()
```

```python
# Modeling

from sklearn.linear_model import LinearRegression # import

lr = LinearRegression() # create an object
lr.fit(X_train, y_train) # fit means study
```

```python
X_test
```

```python
lr.predict(X_test)
```

```python
y_test
```

```python
y_pred = lr.predict(X_test)

list(zip(y_test, y_pred))
```

```python
y_pred_train = lr.predict(X_train)
y_pred_test = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print(mean_absolute_error(y_train, y_pred_train))
print(mean_absolute_error(y_test, y_pred_test))
```

```python
print(mean_squared_error(y_train, y_pred_train))
print(mean_squared_error(y_test, y_pred_test))
```

```python
r2_score_train = r2_score(y_train, y_pred_train)
r2_score_test = r2_score(y_test, y_pred_test)
print(r2_score_train)
print(r2_score_test)
```

```python
# Adjusted R2 Train
N, p = X_train.shape
# print(N)
# print(p)

num = (1-r2_score_train)*(N-1)
den = N - p - 1
ar2_score_train = 1 - (num/den)
print(ar2_score_train)
```

```python
# Adjusted R2 Test
N, p = X_test.shape
# print(N)
# print(p)

num = (1-r2_score_test)*(N-1)
den = N - p - 1
ar2_score_test = 1 - (num/den)
print(ar2_score_test)
```

```python
# Plotting the regression line

plt.scatter(X, Y, color='black')
plt.plot(X, lr.predict(X), color='blue')
plt.title('Regression Line')
plt.xlabel('Advertising Budget')
plt.ylabel('Sales')
plt.show()
```
